##Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

##Subplots

```python
t=np.arange(-10,10,0.1)

fun1=np.sin(t)
fun2=np.cos(t)

plt.subplot(1, 2, 1) #the figure has 1 row, 2 columns, and this plot is the first plot.
plt.xlabel("t")
plt.ylabel("$\sin(t)$")
plt.plot(t,fun1,color='red',linestyle='dotted',linewidth='2',label='$\sin(x)$ Function')  #dashed
plt.legend()

plt.subplot(1, 2, 2) #the figure has 1 row, 2 columns, and this plot is the second plot.
plt.xlabel("t")
plt.ylabel("$\sin(t)$")
plt.plot(t,fun2,color='blue',linestyle='dotted',linewidth='2',label='$\cos(x)$ Function')  #dashed
plt.legend() #loc ="lower left" #fontsize = 18 # shadow = True # facecolor = 'yellow'
```

##Title of Subplot

```python
t=np.arange(-10,10,0.1)

fun1=np.sin(t)
fun2=np.cos(t)

plt.subplot(1, 2, 1) #the figure has 1 row, 2 columns, and this plot is the first plot.
plt.title("A $\sin(t)$ plot versus $t$")
plt.xlabel("t")
plt.ylabel("$\sin(t)$")
plt.plot(t,fun1,color='red',linestyle='dotted',linewidth='2',label='$\sin(x)$ Function')  #dashed
plt.legend(loc="upper left")

plt.subplot(1, 2, 2) #the figure has 1 row, 2 columns, and this plot is the second plot.
plt.title("A $\cos(t)$ plot versus $t$")
plt.xlabel("t")
plt.ylabel("$\sin(t)$")
plt.plot(t,fun2,color='blue',linestyle='dotted',linewidth='2',label='$\cos(x)$ Function')  #dashed
plt.legend(loc="upper left")

plt.tight_layout()
```

##Super Title

```python
t=np.arange(-10,10,0.1)

fun1=np.sin(t)
fun2=np.cos(t)
plt.suptitle("Two Plots: $\sin(t)$ versus $t$ and $\cos(t)$ versus $t$")

plt.subplot(1, 2, 1) #the figure has 1 row, 2 columns, and this plot is the first plot.
plt.title("A $\sin(t)$ plot versus $t$")
plt.xlabel("t")
plt.ylabel("$\sin(t)$")
plt.plot(t,fun1,color='red',linestyle='dotted',linewidth='2',label='$\sin(x)$ Function')  #dashed
plt.legend(loc="upper left")

plt.subplot(1, 2, 2) #the figure has 1 row, 2 columns, and this plot is the second plot.
plt.title("A $\cos(t)$ plot versus $t$")
plt.xlabel("t")
plt.ylabel("$\sin(t)$")
plt.plot(t,fun2,color='blue',linestyle='dotted',linewidth='2',label='$\cos(x)$ Function')  #dashed
plt.legend(loc="upper left")
plt.tight_layout()
```

```python
t=np.arange(-10,10,0.1)

fun1=np.sin(t)
fun2=np.cos(t)
plt.suptitle("Two Plots: $\sin(t)$ versus $t$ and $\cos(t)$ versus $t$",size=16, x=0.5,y=1.1)

plt.subplot(1, 2, 1) #the figure has 1 row, 2 columns, and this plot is the first plot.
plt.title("A $\sin(t)$ plot versus $t$")
plt.xlabel("t")
plt.ylabel("$\sin(t)$")
plt.plot(t,fun1,color='red',linestyle='dotted',linewidth='2',label='$\sin(x)$ Function')  #dashed
plt.legend(loc="upper left")

plt.subplot(1, 2, 2) #the figure has 1 row, 2 columns, and this plot is the second plot.
plt.title("A $\cos(t)$ plot versus $t$")
plt.xlabel("t")
plt.ylabel("$\sin(t)$")
plt.plot(t,fun2,color='blue',linestyle='dotted',linewidth='2',label='$\cos(x)$ Function')  #dashed
plt.legend(loc="upper left")
plt.tight_layout()
```

##Text Annotation

```python
#Now suppose we two curves on a single plot and you want to add text on the plots
x1=np.arange(-5,5,0.01)
x2=np.arange(-5,5,0.01)
f1=x1**2
f2=x2**3
plt.plot(x1,f1)
plt.plot(x2,f2)
```

```python
x1=np.arange(-5,5,0.01)
x2=np.arange(-5,5,0.01)
f1=x1**2
f2=x2**3
plt.plot(x1,f1)
plt.plot(x2,f2)
plt.text(4,25,'$y=x^2$')
```

```python
#1st method
x1=np.arange(-5,5,0.01)
x2=np.arange(-5,5,0.01)
f1=x1**2
f2=x2**3
plt.plot(x1,f1)
plt.plot(x2,f2)
plt.text(4,25,'$y=x^2$')
plt.text(-4.5,-100,'$y=x^3$')
```

```python
#2nd method
x1=np.arange(-5,5,0.01)
x2=np.arange(-5,5,0.01)
f1=x1**2
f2=x2**3
plt.plot(x1,f1)
plt.plot(x2,f2)
plt.annotate('$y=x^2$', xy =(4,10),
                xytext =(4, -100), 
                arrowprops = dict(facecolor ='red'))
```

##Errorbar Plots

Syntax: matplotlib.pyplot.errorbar(x, y, yerr=None, xerr=None, fmt=, ecolor=None, elinewidth=None, capsize=None, barsabove=False, lolims=False, uplims=False, xlolims=False, xuplims=False, errorevery=1, capthick=None, \*, data=None, \*\*kwargs)

```python
x=np.arange(0,10,1)
y=np.array((1,3,2,6,5,3,4,5,9,3))
y_err=y*0.3
# plt.errorbar(x,y,y_err)
# plt.errorbar(x,y,y_err,fmt='b.')
# plt.errorbar(x,y,y_err,fmt='b.',capsize=3)
# plt.errorbar(x,y,y_err,fmt='b.',capsize=3,ecolor='r')
plt.errorbar(x,y,y_err,fmt='b.',capsize=3,ecolor='r',elinewidth=1)
plt.xlabel('x')
plt.ylabel('y')
```

##Confidence levels regions

```python
x=np.arange(0,10,1)
y=np.array((1,2,3,2.8,2.4,2,2.3,2.7,2.3,2))
y_err=y*0.1
plt.errorbar(x,y,y_err,fmt='b.',capsize=3,ecolor='r',elinewidth=1)
plt.fill_between(x, y+1.*y_err, y-1.*y_err, facecolor='#F08080', alpha=0.99)
plt.fill_between(x, y+2.*y_err, y-2.*y_err, facecolor='#F08080', alpha=0.59)
plt.fill_between(x, y+3.*y_err, y-3.*y_err, facecolor='#F08080', alpha=0.39)
```
